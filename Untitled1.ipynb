```python

from nltk import FreqDist
from textblob import TextBlob
import pandas as pd
import numpy as np
import os
```

```python
os.getcwd()
```

```python
text_file_index = os.listdir('../text')
```

```python
text_df = pd.DataFrame(dict(er_text = []))
```

```python
for index, file in enumerate(text_file_index):
    text_df = text_df.append(pd.read_csv('../text/' + file)[['er_text']], ignore_index=True)
```

```python
def replace_punctuation(text):
    import string
    for punc in string.punctuation:
        text = text.replace(punc, ' ')
    text = text.replace('\xa0', ' ')
    return text
```

```python
text_df['without_punc'] = text_df['er_text'].apply(lambda text: replace_punctuation(text))
```

```python
text_df['without_punc'][20]
```

```python
text_df['nouns'] = text_df['er_text'].apply(lambda text: TextBlob(text).noun_phrases)
```

```python
text_df['freq_dic'] = text_df['nouns'].apply(lambda word_ls: FreqDist(word_ls))
```

```python
def freq_dic_agg(arg):
    s = FreqDist()
    for index, item in enumerate(arg):
        if index % 100 == 0:
            print(index)
        s = s + item
    return s
```

```python
len(text_df['freq_dic'].tolist())
```

```python
s = freq_dic_agg(text_df['freq_dic'].tolist())
```

```python
pd.Series(s).sort_values(ascending = False)
```

```python
t = text_df.groupby(['_aux'])['freq_dic'].apply(freq_dic_agg)
```
